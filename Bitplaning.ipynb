```
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
from matplotlib import pyplot as plt
from bitplaning import bitplane
```

```
img = cv2.imread('bitplaning/test2.png',0)
plt.imshow(img,cmap='gray')
plt.show()
```

```
img_in_1D = bitplane.int_to_bit_array(img)


fig,ax = plt.subplots(nrows=2,ncols=4,figsize=(20,10))

for i,axis in enumerate(ax.flat):
  bitplaned_img_in_1D = bitplane.bit_plane(img_in_1D,[i])
  bitplaned_img_in_2D = np.reshape(bitplaned_img_in_1D,img.shape)
  axis.imshow(bitplaned_img_in_2D,cmap='gray')
  axis.set_title(f'Bitplaned along bit - {i+1}')
  axis.set_xticklabels([])
  axis.set_yticklabels([])
plt.show()
```
